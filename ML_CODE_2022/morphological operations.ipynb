```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
blank_img = np.zeros((600,600))
```

```python
blank_img
```

```python
plt.imshow(blank_img, cmap="gray")
```

```python
img = cv2.putText(
    blank_img,
    text = 'ABCDE',
    org = (50,300),
    fontFace = cv2.FONT_HERSHEY_SIMPLEX,
    fontScale=5,
    color=(255,255,255),
    thickness = 25
)
```

```python
plt.imshow(img, cmap='gray')
```

```python
kernal = np.ones((5,5))
```

```python
kernal
```

```python
erode_img = cv2.erode(img,kernal,iterations=4)
```

```python
plt.imshow(erode_img, cmap='gray')
```

```python
dilated_img = cv2.dilate(erode_img,kernal,iterations=1)
```

```python
plt.imshow(dilated_img, cmap='gray')
```

```python
plt.imshow(img , cmap='gray')
```

```python
open_img = cv2.morphologyEx( img ,  cv2.MORPH_OPEN ,kernal )
```

```python
plt.imshow(open_img, cmap='gray')
```

```python
close_img = cv2.morphologyEx( img ,  cv2.MORPH_CLOSE ,kernal )
```

```python
plt.imshow(close_img, cmap='gray')
```
